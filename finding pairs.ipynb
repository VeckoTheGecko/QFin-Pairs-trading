```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
DATA_FOLDER = "individual_stocks_5yr"
DATA_PATHS = [os.path.join(DATA_FOLDER, file_name) for file_name in os.listdir(DATA_FOLDER)]
```

```python
import statsmodels.tsa.stattools as ts
df1 = pd.read_csv(DATA_PATHS[0])
df1["midheight"] = (df1["low"] + df1["high"]) / 2
for index, df2_path in enumerate(DATA_PATHS[1:]):
    df2 = pd.read_csv(df2_path)
    df2["midheight"] = (df2["low"] + df2["high"]) / 2
    _, p_value, _=ts.coint(df1["midheight"],df2["midheight"])
    print(p_value)
```
