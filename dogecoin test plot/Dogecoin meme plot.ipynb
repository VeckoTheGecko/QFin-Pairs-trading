```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from datetime import datetime
from matplotlib.dates import DateFormatter

## Work on formatting dates correctly
## Add xlabel and ylabel, title and image background
```

```python
def format_date(date):
    return datetime.strptime(date, "%b %d, %Y")#datetime.strftime(, "%Y-%m-%d")

DATA = "dogecoin_data.csv"
data = pd.read_csv(DATA)
data["Date"] = data["Date"].apply(format_date)
data = data.sort_values("Date").reset_index()[-200:]

#mpl.
```

```python
plt.xkcd()
#plt.imshow(img, extent=[min(data["Date"]), max(data["Date"]), 0, max(data["Price"])])

plt.plot(data["Date"],data["Price"], color="#e1b303", label = "mega shtonk")





plt.title("The Doge")
plt.xlabel("Date")
plt.ylabel("USD")
plt.legend()

plt.gcf().autofmt_xdate()
plt.savefig("doge.png")
```
